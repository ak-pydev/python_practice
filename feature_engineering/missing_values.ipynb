# Imputation techniques: Missing Value

1. Mean value imputation

```python
sns.histplot(df['age'],kde=True)
```

```python
df['Age_Mean'] = df['age'].fillna(df['age'].mean())
```

```python
sns.histplot(df['Age_Mean'],kde=True)
```

```python
df[['Age_Mean','age']]
```

Median value Impudation - If we have outliers in the dataset

```python
df['Age_Median'] = df['age'].fillna(df['age'].median())
```

```python
sns.histplot(df['Age_Median'],kde=True)
```

Mode Imputation -- Categorical Variables

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['embarked'].notna()]
```

```python
mode_value = df[df['embarked'].notna()]['embarked'].mode()[0] # finding mode value for a column
```

```python
df['embarked_mode']= df['embarked'].fillna(mode_value) # filling categorical NaN values with mode value
```

```python
df[['embarked','embarked_mode']]
```
