# USING MATPLOTPLIB

```python
x  = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Square values")
plt.title("Basic line Plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

#cutomized line plot

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth = 2, markersize = 9)
plt.grid(True)
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
y1 = [1,2,3,4,5]
x1= [12,15,17,19,21]

plt.figure(figsize = (9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='Gold')

plt.subplot(2,2,2)
plt.plot(y,y1,color='Green')

plt.subplot(2,2,3)
plt.plot(y,y1,color='Blue')

plt.subplot(2,2,4)
plt.plot(y1,x1,color='teal')
```

```python
# Bar Plot

categories = ['A','B','C','D','E']
values = [3,6,2,8,1]

plt.bar(categories,values, color ='skyblue')
```

```python
#Histogram

data  = [1,2,2,2,3,3,3,3,3,4,4,1,1,1,1,1,4,4,4,4,5,5,5,5]
plt.hist(data,bins=5,color='lightblue',edgecolor='black')
```

```python
#scatter plot

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.scatter(x,y,color = 'blue',marker='*')
```

```python
#pie char

labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','skyblue']
explode = (0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct = "%1.1f%%",shadow=True)
```

# Using Seaborn

```python
tips= sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips)
plt.title("Scatterplot Plot of Total Bill vs Tip")
```

```python
#line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total bill by size")
plt.show()
```

```python
#Categorical Plots
#Bar plot
sns.barplot(x='day',y='total_bill',data = tips)
plt.title("Barplot of total bills per day")
plt.show()
```

```python
#Box plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Vioilin Plot

sns.violinplot(x= 'day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.pairplot(tips)
```

```python
#heatmap

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='magma')
```

```python
# Analyzing sales data using seaborn

sales_df = pd.read_csv("train.csv")
sales_df.head(5)
```

```python
#Plot total sales by Category
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Sales',data= sales_df,estimator=sum)
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales Price")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Sub-Category',y='Sales',data= sales_df,estimator=sum)
plt.title("Total Sales by Sub Category")
plt.xlabel("Sub Category")
plt.ylabel("Sales Price")
plt.show()
```
