# Naive Baye's Algorithm Practical 

- Binary and Multiclass classification

## Varients of Naive Bayes

- Bernoulli Naive Bayes -> follows bernoulli distribution and features are 0 , 1. 
- Multinomian Naive Bayes -> example Spam Text classification involves use of NLP
- Gaussian Naive Bayes -> follows gaussian distribution and features are continuos

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix,ConfusionMatrixDisplay
```

```python
print(f"Accuracy score: {accuracy_score(y_pred,y_test)}")
```

```python
print("Confusion Matrix: \n\n", confusion_matrix(y_pred,y_test) )
cm = confusion_matrix(y_pred,y_test)
```

```python
print("Classification Report: \n\n", classification_report(y_pred,y_test) )
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Gaussian NB — Confusion Matrix")
plt.show()
```

```python
# side by side comparision
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(range(len(y_test)), y_test, label="Actual", alpha=0.7)
plt.scatter(range(len(y_pred)), y_pred, label="Predicted", alpha=0.7)
plt.legend()
plt.title("Actual vs Predicted Class Labels")
plt.xlabel("Sample Index")
plt.ylabel("Class")
plt.show()

```
