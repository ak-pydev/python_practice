# Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
# creating dataset
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=15)
```

```python
df = pd.DataFrame(X)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
# Model Training

lgm = LogisticRegression()
lgm_model = lgm.fit(X_train,y_train)
```

```python
y_pred = lgm.predict(X_test)
```

```python
y_pred
```

```python
lgm.predict_proba(X_test) # Checking probability
```

```python
# Performance metrics

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

score = accuracy_score(y_test,y_pred)
print(score,"\n")
cm = confusion_matrix(y_test,y_pred)
print(cm,"\n")
classification_report_result = classification_report(y_test,y_pred)
print(classification_report_result)
```

# Hyperparameter Tuning and Cross Validation

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred =  grid.predict(X_test)
```

```python
# Performance metrics

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

score = accuracy_score(y_test,y_pred)
print(score,"\n")
cm = confusion_matrix(y_test,y_pred)
print(cm,"\n")
classification_report_result = classification_report(y_test,y_pred)
print(classification_report_result)
```

# Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV

randomizedCVmodel= RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
randomizedCVmodel.fit(X_train,y_train)
```

```python
y_pred_randomized = randomizedCVmodel.predict(X_test)
```

```python
# Performance metrics

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

score = accuracy_score(y_test,y_pred_randomized)
print(score,"\n")
cm = confusion_matrix(y_test,y_pred_randomized)
print(cm,"\n")
classification_report_result = classification_report(y_test,y_pred_randomized)
print(classification_report_result)
```

# Logistic Regression for Multiclass Classification

```python
X, y = make_classification(n_samples=1000, n_features=10,n_informative=3, n_classes=3, random_state=15) #Creating the dataset
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
logistic = LogisticRegression(multi_class="ovr")
logistic.fit(X_train,y_train)
y_pred_ovr = logistic.predict(X_test) 
```

```python
# Performance metrics

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

score = accuracy_score(y_test,y_pred_ovr)
print(score,"\n")
cm = confusion_matrix(y_test,y_pred_ovr)
print(cm,"\n")
classification_report_result = classification_report(y_test,y_pred_ovr)
print(classification_report_result)
```

## Generating and plotting synthetic imabalanced classification dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
# imabalance dataset exxample

X,y=make_classification(n_samples=10000,n_features=2,n_clusters_per_class=1, n_redundant=0,weights=[0.99],random_state=10)
```

```python
X # checkingg
```

```python
Counter(y)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(max_iter=5000)  # increase max_iter for convergence

param_grid = [
    {
        'penalty': ['l2'],
        'C': [100, 10, 1.0, 0.1, 0.01],
        'solver': ['newton-cg', 'lbfgs', 'sag'],
        'class_weight': ['balanced', None]
    },
    {
        'penalty': ['l1'],
        'C': [100, 10, 1.0, 0.1, 0.01],
        'solver': ['liblinear', 'saga'],
        'class_weight': ['balanced', None]
    },
    {
        'penalty': ['elasticnet'],
        'C': [100, 10, 1.0, 0.1, 0.01],
        'solver': ['saga'],
        'l1_ratio': [0.0, 0.5, 1.0],  # elasticnet needs l1_ratio
        'class_weight': ['balanced', None]
    }
]

grid_search = GridSearchCV(model, param_grid, scoring='roc_auc', cv=5, n_jobs=-1, verbose=2)

grid_search.fit(X, y)

print("Best params:", grid_search.best_params_)
print("Best AUC:", grid_search.best_score_)
```

```python
# plot the roc curve for the model
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 50)) 
# Plot dummy model ROC curve with dashed line
ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')

# Plot actual model ROC curve with markers
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# Annotate thresholds on the curve
for fpr, tpr, thr in zip(model_fpr, model_tpr, thresholds):   
    ax.annotate(f'{thr:.2f}', xy=(fpr, tpr), textcoords="offset points", xytext=(0,5), ha='center')

# Axis labels
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')

# Legend and plot show
ax.legend()
plt.show()
```
