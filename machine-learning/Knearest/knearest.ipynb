# K-nearest neighbour 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(
    n_samples=1000, # observations
    n_features = 3, # total featurees
    n_redundant = 1,
    n_classes = 2, # binary target label
    random_state = 999
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5,algorithm='auto')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
print(f"Accuracy score: {accuracy_score(y_pred,y_test)}")
```

```python
print("Confusion Matrix: \n\n", confusion_matrix(y_pred,y_test) )
cm = confusion_matrix(y_pred,y_test)
```

```python
print("Classification Report: \n\n", classification_report(y_pred,y_test) )
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Reds")
plt.title("K nearest neighbour — Confusion Matrix")
plt.show()
```

```python
# side by side comparision
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(range(len(y_test)), y_test, label="Actual", alpha=0.7)
plt.scatter(range(len(y_pred)), y_pred, label="Predicted", alpha=0.7)
plt.legend()
plt.title("Actual vs Predicted Class Labels")
plt.xlabel("Sample Index")
plt.ylabel("Class")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor(n_neighbors=6,algorithm='auto')

regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("R2 score:",r2_score(y_test,y_pred))


```

```python
print("MAE: ",mean_absolute_error(y_test,y_pred))
```

```python
print("MSE", mean_squared_error(y_test,y_pred))
```
